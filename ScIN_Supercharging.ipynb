```python
#Import and initialize pyrosetta
from pyrosetta import *
init()
```

```python
#Import structure
pose = pyrosetta.rosetta.core.import_pose.pose_from_file('2ln3_1model.pdb')
```

```python
#Set Optional Design Parameters
exclude_gly_pro_cys =  input('Exclude gly/pro/cys residues from design? (Y/N)')
residues_to_exclude = input('Enter residue numbers to exclude from design.')
mode = input('Positive or negative charging? (P/N)')
residues_to_exclude.split(',')
residues_to_exclude_int = []
for value in residues_to_exclude:
    integer = int(value)
    residues_to_exclude_int.append(integer)
```

```python
#Function to calculate protein charge
def get_protein_charge(pose):
    protein_charge = 0
    for i in range (1, pose.total_residue()):
        residue = pose.residue(i)
        if residue.name3() == 'LYS':
            protein_charge += 1
        if residue.name3() == 'ARG':
            protein_charge += 1
        if residue.name3() == 'ASP':
            protein_charge -= 1
        if residue.name3() == 'GLU':
            protein_charge -= 1
    return protein_charge
```

```python
#Function to count mutations
def count_mutations (pose, wild_type_pose):
    mutation_count = 0
    for i in range(1,pose.total_residue()):
        residue = pose.residue(i)
        wild_type_residue = wild_type_pose.residue(i)
        if residue.name1() != wild_type_residue.name1():
            mutation_count += 1
    return mutation_count
```

```python
#Import score function and initialize FastRelax
score_fxn = get_score_function()
fast_relax = pyrosetta.rosetta.protocols.relax.FastRelax(score_fxn)
```

```python
#FastRelax input structure
fast_relax.apply(pose)
wt_relaxed_pose = pose.clone()
```

```python
residue_index_selector = pyrosetta.rosetta.core.select.residue_selector.ResidueIndexSelector()
```

```python
#Identify surface residues
import numpy
surface_neighbor_threshold = 16
surface_residue_indices = []
for i in range(1, pose.total_residue()+1):
    residue_index_selector.set_index(i)
    neighbor_selector = pyrosetta.rosetta.core.select.residue_selector.NeighborhoodResidueSelector(residue_index_selector, 10, False)
    neighbor_count = numpy.sum(neighbor_selector.apply(pose))
    if neighbor_count < surface_neighbor_threshold:
        surface_residue_indices.append(i)
surface_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueIndexSelector()
for index in surface_residue_indices:
    if index not in  residues_to_exclude_int:
        surface_residues.append_index(index)
    
not_surface = pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(surface_residues)
```

```python
#Set up selectors to select surface KRNQ residues for negative charging mode
lys_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('LYS')
arg_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('ARG')
asn_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('ASN')
gln_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('GLN')
pos_residues = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(lys_residues, arg_residues)
neutral_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(asn_residues, gln_residues)
neg_mode_mutatable_residues =  pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(pos_residues, neutral_mutatable_residues)
neg_mode_surface_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(neg_mode_mutatable_residues, surface_residues)    
neg_mode_not_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(neg_mode_surface_mutatable_residues)
surface_lys = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(lys_residues, surface_residues)
surface_arg = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(arg_residues, surface_residues)
surface_asn = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(asn_residues, surface_residues)
surface_gln = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(gln_residues, surface_residues)
```

```python
#Set up selectors to select surface DENQ residues for positive charging mode
asp_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('ASP')
glu_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('GLU')
asn_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('ASN')
gln_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('GLN')
neg_residues = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(asp_residues, glu_residues)
neutral_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(asn_residues, gln_residues)
pos_mode_mutatable_residues =  pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(neg_residues, neutral_mutatable_residues)
pos_mode_surface_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(pos_mode_mutatable_residues, surface_residues)    
pos_mode_not_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(pos_mode_surface_mutatable_residues)
surface_asp = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(asp_residues, surface_residues)
surface_glu = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(glu_residues, surface_residues)
surface_asn = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(asn_residues, surface_residues)
surface_gln = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(gln_residues, surface_residues)
```

```python
#Set up task factory to design KR to DENQST and NQ to DE and pack all other residues for negative mode
tf1_neg = pyrosetta.rosetta.core.pack.task.TaskFactory()
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.InitializeFromCommandline())
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.IncludeCurrent())

# Disable design on not_mutatable_residues
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    pyrosetta.rosetta.core.pack.task.operation.RestrictToRepackingRLT(), neg_mode_not_mutatable_residues))

# Enable design on surface_lys residues
aa_to_design_lys = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_lys.aas_to_keep("DENQST")
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_lys, surface_lys))

# Enable design on surface_arg residues
aa_to_design_arg = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_arg.aas_to_keep("DENQST")
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_arg, surface_arg))

# Enable design on surface_asn residues
aa_to_design_asn = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_asn.aas_to_keep("NDE")
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_asn, surface_asn))

# Enable design on surface_gln residues
aa_to_design_gln = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_gln.aas_to_keep("QDE")
tf1_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_gln, surface_gln))

# Convert the task factory into a PackerTask
neg_packer_task1 = tf1_neg.create_task_and_apply_taskoperations(wt_relaxed_pose)
# View the PackerTask
print(neg_packer_task1)
```

```python
#Set up task factory to design DE to KRNQST and NQ to KR and pack all other residues for positive mode
tf1_pos = pyrosetta.rosetta.core.pack.task.TaskFactory()
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.InitializeFromCommandline())
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.IncludeCurrent())

# Disable design on not_mutatable_residues
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    pyrosetta.rosetta.core.pack.task.operation.RestrictToRepackingRLT(), pos_mode_not_mutatable_residues))

# Enable design on surface_lys residues
aa_to_design_asp = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_asp.aas_to_keep("KRNQST")
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_asp, surface_asp))

# Enable design on surface_arg residues
aa_to_design_glu = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_glu.aas_to_keep("KRNQST")
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_glu, surface_glu))

# Enable design on surface_asn residues
aa_to_design_asn = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_asn.aas_to_keep("NKR")
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_asn, surface_asn))

# Enable design on surface_gln residues
aa_to_design_gln = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASRLT()
aa_to_design_gln.aas_to_keep("QKR")
tf1_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_gln, surface_gln))

# Convert the task factory into a PackerTask
pos_packer_task1 = tf1_pos.create_task_and_apply_taskoperations(wt_relaxed_pose)
# View the PackerTask
print(pos_packer_task1)
```

```python
#Set Up Move Map for FastDesign
mm = pyrosetta.rosetta.core.kinematics.MoveMap()
mm.set_bb(True)
mm.set_chi(True)
mm.set_jump(True)
```

```python
#Construct fast_design1 using tf1 and Move Map
fast_design1 = pyrosetta.rosetta.protocols.denovo_design.movers.FastDesign(scorefxn_in=score_fxn, standard_repeats=10)
if mode == 'P':
    fast_design1.set_task_factory(tf1_pos)
if mode =='N':
    fast_design1.set_task_factory(tf1_neg)
fast_design1.set_movemap(mm)
fast_design1.minimize_bond_angles(False)
fast_design1.minimize_bond_lengths(False)
```

```python
pose.assign(wt_relaxed_pose)
fast_design1.apply(pose)
```

```python
wt_relaxed_score = score_fxn(wt_relaxed_pose)
fast_design1_output_pose = pose.clone()
fd1_pose_score = score_fxn(fast_design1_output_pose)
fd1_ddg = fd1_pose_score - wt_relaxed_score
fd1_charge = get_protein_charge(fast_design1_output_pose)
fd1_mutations = count_mutations(fast_design1_output_pose, wt_relaxed_pose)
fd1_rmsd = pyrosetta.rosetta.core.scoring.CA_rmsd(fast_design1_output_pose,  wt_relaxed_pose)
print('ddg:', fd1_ddg)
print('Charge:', fd1_charge)
print('# of mutations:', fd1_mutations)
print('RMSD:', fd1_rmsd)
```

```python
#Set up residue selectors for second round of design
gly_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('GLY')
pro_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('PRO')
cys_residues = pyrosetta.rosetta.core.select.residue_selector.ResidueNameSelector('CYS')
gly_pro = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(gly_residues, pro_residues)
gly_pro_cys = pyrosetta.rosetta.core.select.residue_selector.OrResidueSelector(gly_pro, cys_residues)
not_gly_pro_cys = pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(gly_pro_cys)
if exclude_gly_pro_cys == 'Y':
    design_2_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.AndResidueSelector(surface_residues, not_gly_pro_cys)
    design_2_not_mutatable_residues =  pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(design_2_mutatable_residues)
if exclude_gly_pro_cys =='N':
    design_2_mutatable_residues = surface_residues
    design_2_not_mutatable_residues = pyrosetta.rosetta.core.select.residue_selector.NotResidueSelector(design_2_mutatable_residues)
```

```python
#Set up task factory to design surface residues to DE for negative charging
tf2_neg = pyrosetta.rosetta.core.pack.task.TaskFactory()
tf2_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.InitializeFromCommandline())
tf2_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.IncludeCurrent())

# Disable design on not_surface residues
tf2_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    pyrosetta.rosetta.core.pack.task.operation.RestrictToRepackingRLT(), design_2_not_mutatable_residues))

# Enable design on surface_residues
aa_to_design_surface_neg = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASExceptNativeRLT()
aa_to_design_surface_neg.aas_to_keep("DE")
tf2_neg.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_surface_neg, design_2_mutatable_residues))

# Convert the task factory into a PackerTask
neg_packer_task2 = tf2_neg.create_task_and_apply_taskoperations(fast_design1_output_pose)
# View the PackerTask
print(neg_packer_task2)
```

```python
#Set up task factory to design surface residues to KR for positive charging
tf2_pos = pyrosetta.rosetta.core.pack.task.TaskFactory()
tf2_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.InitializeFromCommandline())
tf2_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.IncludeCurrent())

# Disable design on not_surface residues
tf2_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    pyrosetta.rosetta.core.pack.task.operation.RestrictToRepackingRLT(), design_2_not_mutatable_residues))

# Enable design on surface_residues
aa_to_design_surface_pos = pyrosetta.rosetta.core.pack.task.operation.RestrictAbsentCanonicalAASExceptNativeRLT()
aa_to_design_surface_pos.aas_to_keep("KR")
tf2_pos.push_back(pyrosetta.rosetta.core.pack.task.operation.OperateOnResidueSubset(
    aa_to_design_surface_pos, design_2_mutatable_residues))

# Convert the task factory into a PackerTask
pos_packer_task2 = tf2_pos.create_task_and_apply_taskoperations(fast_design1_output_pose)
# View the PackerTask
print(pos_packer_task2)
```

```python
fast_design2 = pyrosetta.rosetta.protocols.denovo_design.movers.FastDesign(scorefxn_in=score_fxn, standard_repeats=10)
if mode == 'P':
    fast_design2.set_task_factory(tf2_pos)
if mode == 'N':
    fast_design2.set_task_factory(tf2_neg)
fast_design2.set_movemap(mm)
fast_design2.minimize_bond_angles(False)
fast_design2.minimize_bond_lengths(False)
```

```python
pose.assign(fast_design1_output_pose)
fast_design2.apply(pose)
```

```python
fast_design2_output_pose = pose.clone()
fd2_pose_score = score_fxn(fast_design2_output_pose)
fd2_ddg = fd2_pose_score - wt_relaxed_score
fd2_charge = get_protein_charge(fast_design2_output_pose)
fd2_mutations = count_mutations(fast_design2_output_pose, wt_relaxed_pose)
fd2_rmsd = pyrosetta.rosetta.core.scoring.CA_rmsd(fast_design2_output_pose,  wt_relaxed_pose)
print('ddg:', fd2_ddg)
print('Charge:', fd2_charge)
print('# of mutations:', fd2_mutations)
print('RMSD:', fd2_rmsd)
```
